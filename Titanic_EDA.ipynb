1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Load Dataset

```python
df = pd.read_csv(r"C:\Users\varshini\Downloads\train.csv")
```

4. Survival Distribution

```python
sns.countplot(x='Survived', data=df)
plt.show()
```

5. Passenger Class Analysis

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.show()
```

6. Age Distribution

```python
plt.hist(df['Age'].dropna(), bins=20)
plt.show()
```

7. Boxplot

```python
sns.boxplot(x=df['Fare'])
plt.show()
```

8. Correlation Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')
plt.show()
```

9. Scatterplot

```python
sns.scatterplot(x='Age',
                y='Fare',
                hue='Survived',
                data=df)

plt.show()
```

10. Pairplot

```python
sns.pairplot(df[['Age','Fare','Pclass','Survived']])
plt.show()
```
